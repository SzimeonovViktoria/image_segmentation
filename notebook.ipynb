No_Hemorrhage oszlop cseréje, hogy érthetőbbek legyenek maga az adatok

```python
def swap_target(x):
    if x == 0:
        return 1
    else:
        return 0


df_diag['Has_Hemorrhage'] = df_diag['No_Hemorrhage'].apply(swap_target)
```

régi oszlop törlése

```python
df_diag = df_diag.drop('No_Hemorrhage', axis=1)
print(df_diag.shape)
```

```python
df_diag.head()
```

84.páciens, 36.vizsgálata nem teljes, nincs hozzá kép
=> hozzá tartozó képek törlése

```python
index_to_drop = df_diag[(df_diag['PatientNumber'] == 84) & (df_diag['SliceNumber'] == 36)].index
index_to_drop = index_to_drop[0]

df_diag = df_diag.drop(index_to_drop, axis=0)
```

minden páciens adathoz hozzáadom mi a hozzátartozó kép

```python
def get_mask_fname(row):       
    mask_id = str(row['SliceNumber']) + '_HGE_Seg.jpg'
    return mask_id
    
df_diag['mask_fname'] = df_diag.apply(get_mask_fname, axis=1)
```

```python
def new_mask_fname(row):  
    mask_id = str(row['PatientNumber']) + '_' + str(row['SliceNumber']) + '_HGE_Seg.jpg'
    return mask_id

df_diag['new_mask_fname'] = df_diag.apply(new_mask_fname, axis=1)
```

```python
def assign_image_fname(row):   
    image_fname = str(row['SliceNumber']) + '.jpg'  
    return image_fname

df_diag['image_fname'] = df_diag.apply(assign_image_fname, axis=1)
```

```python
def assign_new_fname(row):        
    mask_id = str(row['PatientNumber']) + '_' + str(row['SliceNumber']) + '.jpg'   
    return mask_id
    
df_diag['new_image_fname'] = df_diag.apply(assign_new_fname, axis=1)
```

adathalmaz felosztása
k-fold

```python
k=1
test_img = list(range(49, 131))

while k<4:
 
    all_patient_number = list(range(49, 131))

    dir = 'fold_' + str(k) + '/test_image_dir'
    for f in os.listdir(dir):
        os.remove(os.path.join(dir, f))

    dir =  'fold_' + str(k) +  '/train_image_dir'
    for f in os.listdir(dir):
        os.remove(os.path.join(dir, f))

    dir =  'fold_' + str(k) + '/val_image_dir'
    for f in os.listdir(dir):
        os.remove(os.path.join(dir, f))


    test_image_dir = 'fold_' + str(k) + '/test_image_dir'

    train_image_dir = 'fold_' + str(k) +  '/train_image_dir'

    val_image_dir = 'fold_' + str(k) + '/val_image_dir'

    #train adathalmaz eloállítása-----------------------------------------------------------------------
    train_images_count = 0

    while train_images_count <= 1500:

        patient_imgs = random.choice(all_patient_number)
        all_patient_number.remove(patient_imgs)

        patient_num = patient_imgs
        df = df_diag[df_diag['PatientNumber'] == patient_num]
        image_list = list(df['image_fname'])
            
        for fname in image_list:
            new_fname = str(patient_num) + '_' + fname
            if(patient_num<100):
                path = base_path + '/Patients_CT/' + '0' + str(patient_num) + '/brain'
            else:
                path = base_path + '/Patients_CT/' + str(patient_num) + '/brain'
            src = os.path.join(path, fname)
            dst = os.path.join(train_image_dir, new_fname)
            shutil.copyfile(src, dst)
        train_images_count+=len(path)



    #test adathalmaz eloállítása------------------------------------------------------------------------
    test_images_count=0

    while test_images_count <= 500:     
        patient_imgs = random.choice(test_img)

        for i in all_patient_number:
            if i==patient_imgs:
                all_patient_number.remove(i)

        for i in test_img:
            if i==patient_imgs:
                test_img.remove(i)
                
        patient_num = patient_imgs
        df = df_diag[df_diag['PatientNumber'] == patient_num]
        image_list = list(df['image_fname'])
            
        for fname in image_list:
            new_fname = str(patient_num) + '_' + fname
            if(patient_num<100):
                path = base_path + '/Patients_CT/' + '0' + str(patient_num) + '/brain'
            else:
                path = base_path + '/Patients_CT/' + str(patient_num) + '/brain'
            src = os.path.join(path, fname)
            dst = os.path.join(test_image_dir, new_fname)
            shutil.copyfile(src, dst)
        test_images_count+=len(path)


    #val adathalmaz eloállítása-----------------------------------------------------------------------
    val_images_count = 0

    while val_images_count <= 500:
        patient_imgs = random.choice(all_patient_number)
        all_patient_number.remove(patient_imgs)

        patient_num = patient_imgs
        df = df_diag[df_diag['PatientNumber'] == patient_num]
        image_list = list(df['image_fname'])
            
        for fname in image_list:
            new_fname = str(patient_num) + '_' + fname
            if(patient_num<100):
                path = base_path + '/Patients_CT/' + '0' + str(patient_num) + '/brain'
            else:
                path = base_path + '/Patients_CT/' + str(patient_num) + '/brain'
            src = os.path.join(path, fname)
            dst = os.path.join(val_image_dir, new_fname)
            shutil.copyfile(src, dst)
        val_images_count+=len(path)
    k+=1
```
